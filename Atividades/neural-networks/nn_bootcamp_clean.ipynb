```python
def scaleFeatures(training_data, numeric_features, date_features):
    ss = StandardScaler()

    # fit scaler on numeric_features
    ss.fit(training_data[numeric_features + date_features])

    # scale numeric features now
    training_data_scale = training_data.copy()
    training_data_scale[numeric_features + date_features] = ss.transform(training_data[numeric_features + date_features])
    
    return [ss, training_data_scale]

def scaleAllFeatures(training_data):
    ss = StandardScaler()

    # fit scaler on numeric_features
    ss.fit(training_data)

    # scale numeric features now
    training_data_scale = training_data.copy()
    training_data_scale = ss.transform(training_data_scale)
    
    return [ss, training_data_scale]

def oneHotEncoding(training_data, categorical_features):
    training_data = pd.get_dummies(training_data, columns=categorical_features)
    return training_data


####TRAINING AND EVALUATION####
def trainClassifier(clf, training_data, training_results, feature=0):
    clf.fit(training_data, np.array(training_results[feature]))
    print("Training finished!")
    #print(clf.feature_importances_)
    prediction_results = clf.predict(training_data);
    print('Accuracy: ', float(accuracy_score(np.array(training_results[feature]), prediction_results)) * 100, '%')
    return clf

def evaluateClassifier(clf, test_data, test_results, feature=0, crossValidate=False):
    prediction_results = clf.predict(test_data);
    print("Evaluation finished!")
    print('Accuracy: ', float(accuracy_score(np.array(test_results[feature]), prediction_results)) * 100, '%')

    print('Classification Stats:')
    print(classification_report(prediction_results, np.array(test_results[feature])))

    if crossValidate:
        scores = cross_val_score(clf, test_data, np.array(test_results[feature]))
        print("Cross validation: ", scores.mean()) 

    plot_all_confusion_matrix(actual_labels=np.array(test_results[feature]), pred_labels=prediction_results)
    
def trainAndEvaluate(clf, training_data, training_results, test_data, test_results, feature=0):
    clf = trainClassifier(clf, training_data, training_results, feature)
    evaluateClassifier(clf, test_data, test_results, feature)
    return clf

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

def plot_all_confusion_matrix(actual_labels, pred_labels, classes=['CLASS_0','CLASS_1']):
    # Compute confusion matrix
    cnf_matrix = confusion_matrix(y_true=actual_labels, y_pred=pred_labels)
    np.set_printoptions(precision=2)

    # Plot non-normalized confusion matrix
    plt.figure()
    plot_confusion_matrix(cnf_matrix, classes=classes,
                         title='Confusion matrix, without normalization')

    # Plot normalized confusion matrix
    plt.figure()
    plot_confusion_matrix(cnf_matrix, classes=classes, normalize=True,
                          title='Normalized confusion matrix')

    plt.show()
    
    

def plotData2D(training_data_pca):
    fig, ax = plt.subplots()
    #sampled = np.random.choice(len(training_data_pca),len(training_data_pca))
    ax.scatter(training_data_pca[:,0], training_data_pca[:,1], c=training_data_pca[:,2])#, )

    ax.set_title('PCA 2Ds')
    plt.show()

def plotOptimalFeatures(estimator, X_train, y_train):
    from sklearn.feature_selection import RFECV

    rfecv = RFECV(estimator=estimator, step=1, cv=5, scoring='accuracy')
    rfecv= rfecv.fit(X_train, y_train)

    print('Optimal number of features: ', rfecv.n_features_)
    print('Best features: ', X_train.columns[rfecv.support_])
    
    import matplotlib.pyplot as plt

    plt.figure()
    plt.xlabel("Number of features selected")
    plt.ylabel("Cross validation score of number of selected features")
    plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
    plt.show()
```
